# FEATURE ENGINEERING: Creating new features from existing features for better analytical insights

# Adding Temporal Features:

```python
#Time-Based Features
bt_master['month_year'] = bt_master['order_date'].dt.to_period('M')
bt_master['order_day'] = bt_master['order_date'].dt.day_name()
bt_master['order_hour'] = bt_master['order_date'].dt.hour
bt_master['is_weekend'] = bt_master['order_date'].dt.weekday >= 5
bt_master['order_quarter'] = bt_master['order_date'].dt.quarter
bt_master['day_of_month'] = bt_master['order_date'].dt.day
bt_master['week_of_year'] = bt_master['order_date'].dt.isocalendar().week
```

# Adding Delivery Features:

```python
#Delivery Delay (Minutes)
bt_delp['promised_time'] = pd.to_datetime(bt_delp['promised_time'], errors='coerce')
bt_delp['actual_time'] = pd.to_datetime(bt_delp['actual_time'], errors='coerce')

bt_delp['delivery_delay'] = (bt_delp['actual_time'] - bt_delp['promised_time']).dt.total_seconds() / 60

bt_master = pd.merge(
    bt_master,
    bt_delp[['order_id', 'delivery_delay']],
    on='order_id',
    how='left'
)
```

# Adding Customer Features:

```python
#Feature 2: Order Value per Customer
#Computing customer-wise order value potential
bt_cust['order_value_per_customer'] = bt_cust['total_orders'] * bt_cust['avg_order_value']

bt_master = pd.merge(
    bt_master,
    bt_cust[['customer_id', 'order_value_per_customer']],
    on='customer_id',
    how='left'
)
```

```python
#Inventory Summary
bt_invtnew['date'] = pd.to_datetime(bt_invtnew['date'], errors='coerce')

# Aggregate inventory by product
inv_summary = bt_invtnew.groupby('product_id')[['stock_received', 'damaged_stock']].sum().reset_index()

bt_master = pd.merge(bt_master, inv_summary, on='product_id', how='left')
```

# Adding Product-related features:

```python
#Total Item Value
bt_master['total_item_value'] = bt_master['quantity'] * bt_master['unit_price']

#Available Stock
bt_master['available_stock'] = bt_master['stock_received'] - bt_master['damaged_stock']

#Stock Sufficieny ratio: This ratio will indicate whether the available stock was enough to fulfill the order
#(values >= 1 mean sufficient stock).
bt_master['stock_sufficiency_ratio'] = bt_master['available_stock'] / bt_master['quantity']
```

# 3. Weekly Sales Trend

```python
#Weekly Sales Trend

# Extract ISO year and week
bt_master['iso_year'] = bt_master['order_date'].dt.isocalendar().year
bt_master['iso_week'] = bt_master['order_date'].dt.isocalendar().week

# WEEKLY revenue = sum of total_item_value (correct product-level revenue)
weekly_sales = (
    bt_master.groupby(['iso_year', 'iso_week'])
    .agg(
        total_sales=('total_item_value', 'sum'),
        total_orders=('order_id', 'nunique')
    )
    .reset_index()
    .sort_values(['iso_year', 'iso_week'])
)

# Create a week-start date (best for plotting time series)
weekly_sales['week_start'] = pd.to_datetime(
    weekly_sales['iso_year'].astype(str) +
    weekly_sales['iso_week'].astype(str) +
    '1',
    format='%G%V%u'  # ISO year-week-weekday
)

#Weekly Total Sales Trend
plt.figure(figsize=(12, 5))
sns.lineplot(
    x='week_start',
    y='total_sales',
    data=weekly_sales,
    marker='o',
    color='teal'
)
plt.title('Weekly Sales Trend', fontsize=14, fontweight='bold')
plt.ylabel('Total Weekly Sales ()')
plt.xlabel('Week')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

# TIME SERIES FORECASTING (Predictive Modelling):
**Objective:** The main goal is to forecast future product demand (weekly basis) for better demand planning and supply chain optimization.
Here, we are using:
1.	ARIMA (AutoRegressive Integrated Moving Average)
2.	Exponential Smoothing (Holt-Winters)


```python
#Data Preparation for Time-Series Forecasting
#aggregate daily sales totals

bt_master['order_date'] = pd.to_datetime(bt_master['order_date'], errors='coerce')

# Aggregate sales by date
daily_sales = bt_master.groupby('order_date')['order_total'].sum().reset_index()
daily_sales = daily_sales.sort_values('order_date')
daily_sales.set_index('order_date', inplace=True)

print(daily_sales.head())
```

```python
#STATIONARY CHECK (ADF Test)
#Before applying ARIMA, we check if the time series is stationary (constant mean and variance).

from statsmodels.tsa.stattools import adfuller

# Perform Augmented Dickey-Fuller Test
adf_result = adfuller(daily_sales['order_total'])

print("ADF Statistic:", adf_result[0])
print("p-value:", adf_result[1])

if adf_result[1] < 0.05:
    print("Data is stationary. We can proceed (good for ARIMA)")
else:
    print("Data is non-stationary. Differencing needed")
```

```python
# Aggregate daily sales to weekly sales (sum of order totals)
weekly_sales = daily_sales['order_total'].resample('W').sum()
print(weekly_sales.head())
```

```python
#Stationarity Check on Weekly Sales
from statsmodels.tsa.stattools import adfuller

adf_test = adfuller(weekly_sales)
print("ADF Statistic:", adf_test[0])
print("p-value:", adf_test[1])

if adf_test[1] < 0.05:
    print("Weekly sales data is stationary. ARIMA can be applied directly")
else:
    print("Weekly sales data is non-stationary. Differencing or transformations may be required before ARIMA")
```

# 1. ARIMA Modeling & 6-Month Weekly Forecast

```python
#ARIMA Modeling & 6-Month Weekly Forecast
from statsmodels.tsa.arima.model import ARIMA

#Fit ARIMA model on weekly sales data
model = ARIMA(weekly_sales, order=(1,1,1))
model_fit = model.fit()

# --- Forecast next 26 weeks (~6 months) ---
forecast_steps = 26
forecast = model_fit.get_forecast(steps=forecast_steps)

# Extract predicted mean and confidence intervals
forecast_mean = forecast.predicted_mean
conf_int = forecast.conf_int()

# Create forecast DataFrame
forecast_index = pd.date_range(
    start=weekly_sales.index[-1] + pd.Timedelta(weeks=1),
    periods=forecast_steps,
    freq='W'
)

forecast_df_weekly_6m = pd.DataFrame({
    'week_start_date': forecast_index,
    'forecasted_sales': forecast_mean.values,
    'lower_ci': conf_int.iloc[:, 0].values,
    'upper_ci': conf_int.iloc[:, 1].values
})

#Save forecast to CSV
forecast_df_weekly_6m.to_csv("weekly_sales_forecast_6months.csv", index=False)

print("6-Month Weekly Sales Forecast saved as 'weekly_sales_forecast_6months.csv")
print(forecast_df_weekly_6m.head())
```

```python
#Plot Forecast
plt.figure(figsize=(12,6))
plt.plot(weekly_sales.index, weekly_sales, label='Historical Weekly Sales', color='blue')
plt.plot(forecast_index, forecast_mean, label='Forecasted Weekly Sales (Next 6 Months)', color='red')
plt.fill_between(forecast_index, conf_int.iloc[:, 0], conf_int.iloc[:, 1],
                 color='pink', alpha=0.3)
plt.title(" Weekly Sales Forecast for Next 6 Months (ARIMA Model)")
plt.xlabel("Week")
plt.ylabel("Weekly Sales ()")
plt.legend()
plt.show()
```

**Observation:** ARIMA model predicts no significant growth or decline. It expects weekly sales to stabilize around the historical average.

# 2. Exponential Smoothing Model

```python
#Holt-Winters (Exponential Smoothing) Weekly Forecast
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Fit Holt-Winters model (trend + seasonality)
hw_model = ExponentialSmoothing(weekly_sales, trend='add', seasonal='add', seasonal_periods=12).fit()

#Forecast Next 26 Weeks (~6 Months)
hw_forecast = hw_model.forecast(steps=26)

# Plot
plt.figure(figsize=(12,6))

#Plot historical weekly sales
plt.plot(weekly_sales, label='Historical Weekly Sales', color='blue')

## Plot Holt-Winters forecast
plt.plot(hw_forecast, label='Holt-Winters Forecast (6 Months)', color='red')

plt.title('Weekly Sales Forecast - Exponential Smoothing (Holt Winters) Model')
plt.xlabel('Week')
plt.ylabel('Weekly Sales ()')
plt.legend()
plt.show()
```
